```python
col = []
for i in data.find("table"):
    # print(i.find('title'))
    for j in i.find_all('tr'):
        for k in j.find_all('th'):
            for l in k:
                col.append(l.text)
col   
```

```python
import pandas as pd
data1 = pd.DataFrame(columns= col)
data1
```

```python
for row in data.find('tbody').find_all('tr'):
    colz = row.find_all('td')
    name= colz[0].text
    las = colz[1].text
    chg= colz[2].text
    vol = colz[3].text
    # data1 = data1.append
    new = pd.DataFrame({col[0]:name,col[1]:las,col[2]:chg,col[3]:vol},index=[0])
    # print(new)
    data1 =pd.concat([data1, new]).reset_index(drop=True)
data1
```
